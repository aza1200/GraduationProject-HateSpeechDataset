```python
import os
import koco
import json
import requests
import pandas as pd
from datasets import load_dataset
```

# Apeach download

```python
# 1. apeach downalod
apeach_dataset_name = "jason9693/APEACH"

apeach_dataset = load_dataset(apeach_dataset_name)

apeach_train_df =  pd.DataFrame(apeach_dataset['train'])
apeach_test_df = pd.DataFrame(apeach_dataset['test'])
apeach_df = pd.concat([apeach_train_df, apeach_test_df], ignore_index=True)
apeach_df.to_excel('./apeach.xlsx',index=False)
apeach_df.sample(10)
```

# Hate Score Download

```python
hatescore_dataset = load_dataset('smilegate-ai/kor_unsmile')
```

```python
text_list = []
label_list = []
for tmp_dict in hatescore_dataset['train']:
  text_list.append(tmp_dict['문장'])
  tmp_label = []
  for tmp_key in list(tmp_dict.keys()):
    if tmp_key != '문장' and tmp_dict[tmp_key] == 1:
      tmp_label.append(tmp_key)
  label_list.append(tmp_label)

for tmp_dict in hatescore_dataset['valid']:
  text_list.append(tmp_dict['문장'])
  tmp_label = []
  for tmp_key in list(tmp_dict.keys()):
    if tmp_key != '문장' and tmp_dict[tmp_key] == 1:
      tmp_label.append(tmp_key)
  label_list.append(tmp_label)
```

```python
hatescore_df = pd.DataFrame({
    'text' : text_list,
    'class': label_list
})

hatescore_df.to_excel('./hatescore.xlsx',index=False)
hatescore_df.sample(10)
```

# K-MHAS download

```python
kmhas_dataset = load_dataset("jeanlee/kmhas_korean_hate_speech")
```

```python
train_kmhas_df = pd.DataFrame(kmhas_dataset['train'])
test_kmhas_df = pd.DataFrame(kmhas_dataset['test'])
valid_kmhas_df = pd.DataFrame(kmhas_dataset['validation'])

kmhas_df = pd.concat([train_kmhas_df, test_kmhas_df, valid_kmhas_df], ignore_index=True)
kmhas_df.to_excel('./kmhas.xlsx',index=False)
kmhas_df.sample(10)
```
