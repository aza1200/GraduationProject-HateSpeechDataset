```python
import pandas as pd

# 일반 발언 0
all_df = pd.DataFrame(columns=["text","class"])

apeach_df = pd.read_excel('./apeach.xlsx')
beep_df = pd.read_excel('./beep.xlsx')
hate_score_df = pd.read_excel('./hate_score.xlsx')
KMHAS_df = pd.read_excel('./kmhas.xlsx')
kold_df = pd.read_excel('./kold.xlsx')
```

```python
# apeach_df
# 건너뜀
# hate 와 non-hate 로만 분류되어있음 

# non hate 가져올수 있을듯 

common_speach_apeach_series = apeach_df[apeach_df['class'] == 0]['text']
# common_speach_beep_df = beep_df[beep_df['hate']=='none']['comments']
# common_speach_hate_score_df = hate_score_df[hate_score_df['macrolabel']=='Clean']['comment']
# common_speach_KMHAS_df = KMHAS_df[KMHAS_df['label']=='[8]']['text']
# common_speach_kold_df = kold_df[kold_df['OFF']==False]['comment']

# common_series_all = pd.concat([
#     common_speach_apeach_df,
#     common_speach_beep_df,
#     common_speach_hate_score_df,
#     common_speach_KMHAS_df,
#     common_speach_kold_df
# ], ignore_index=True)

labeled_apeach_df = pd.DataFrame(common_speach_apeach_series, columns=["text"])

# 모든 행에 대해 [0]을 값으로 가지는 리스트 생성
class_list = [[0] for _ in range(len(labeled_apeach_df))]

# 'class' 컬럼에 생성된 리스트 할당
labeled_apeach_df['class'] = class_list
print(len(labeled_apeach_df))
labeled_apeach_df.head()
```

```python
# beep_df

# 성차별 
# 일반 발언 
# 이렇게 나눌수있음 
# beep_sex_discrimination_series = beep_df[
#     (beep_df['contain_gender_bias'] ==  True) & (beep_df['hate'].isin(['hate', 'offensive']))
# ]['comments']

beep_sex_discrimination_series = beep_df[
    (beep_df['contain_gender_bias'] ==  True) & (beep_df['hate'].isin(['hate', 'offensive']))
]['comments']
common_speach_beep_seires = beep_df[beep_df['hate']=='none']['comments']


sex_beep_df = pd.DataFrame(beep_sex_discrimination_series, columns=["comments"])
common_beep_df = pd.DataFrame(common_speach_beep_seires, columns=["comments"])


class_list_sex = [[1] for _ in range(len(sex_beep_df))]
class_list_common = [[0] for _ in range(len(common_beep_df))]


sex_beep_df["class"] = class_list_sex
common_beep_df["class"] = class_list_common

sex_beep_df.rename(columns={'comments': 'text'}, inplace=True)
common_beep_df.rename(columns={'comments': 'text'}, inplace=True)

labeled_beep_df = pd.concat([sex_beep_df, common_beep_df], ignore_index=True)
print(len(labeled_beep_df))
labeled_beep_df.head()
```

```python
hate_score_df.head()
```

```python
# None -> 0
# 지역 -> 7
# 종교 -> 3
# 인종/국적 -> 2
# 연령 -> 4
# 성소수자 -> 8
# 단순 악플 -> 6
# 남성 + # 여성/가족 -> 1
# 기타 혐오 -> 8
class_change_dict = {
    "지역": 7, "종교":3, "인종/국적":2,
    "연령":4, "성소수자":9, "악플/욕설":6, "남성":1, "여성/가족": 1,
    "기타 혐오":10,'clean':0
}

class_list = []
for idx in range(len(hate_score_df)):
    labels = eval(hate_score_df.loc[idx]['class'])
    tmp_class = []
    for label in labels:
        if label == '개인지칭': continue
        after_change = class_change_dict[label]
        tmp_class.append(after_change)
    class_list.append(tmp_class)

hate_score_df["class"] = class_list
hate_score_df.rename(columns={'comment': 'text'}, inplace=True)

labeled_hate_score_df = hate_score_df[["text","class"]]
labeled_hate_score_df.head()
```

```python
# KMHAS_df
# 0: 출신
# 1: 외모
# 2: 정치성향
# 3 : 혐오욕설
# 4 : 연령 차별
# 5 : 성차별
# 6: 인종차별 
# 7 : 종교차별
# 8 : 일반 발언 -> 0
class_change_dict = {
    "0":7 ,'1':8 , '2':5, '3':6,
    '4':4, '5':1 , '6':2, '7':3,
    '8':0
}
class_list = []
for idx in range(len(KMHAS_df)):
    labels = KMHAS_df.loc[idx].label.replace('[','').replace(']','').split(", ")
    tmp_class = []
   
    for label in labels:
        after_change = class_change_dict[label]
        tmp_class.append(after_change)
    class_list.append(tmp_class)

KMHAS_df["class"] = class_list
labeled_KMHAS_df = KMHAS_df[["text","class"]]
print(len(labeled_KMHAS_df))
labeled_KMHAS_df.head()
```

```python
# 페미관련 발언들 -> other ?
kold_df.head()

# offensive가 false 면 일반발언 0
# - 앞에 gender 들어가면 성차별로 (1)
# - 앞에 race 들어가면 인종 (2)
# - 앞에 religion 들어가면 종교 (3)
# - 앞에 politics 들어가면 정치 (5)
# other-age 연령차별 (4)
# other-feminist 성차별 (1)

class_list = []
text_list = []

for idx in range(len(kold_df)):
    comment = kold_df.loc[idx].comment
    grps = kold_df.loc[idx].GRP
    is_offensive = kold_df.loc[idx].OFF
    tmp_class = []
    will_append = False
    if is_offensive == False:
        tmp_class.append(0)
        will_append = True
    else:
        if pd.isna(kold_df.loc[idx, 'GRP']) == True:
            continue
        labels = grps.split(" & ")

        for label in labels:
            first = label.split("-")[0]
            second = label.split("-")[1]
            
            if first == "gender":
                tmp_class.append(1)
                will_append = True
            elif first == "race":
                tmp_class.append(2)
                will_append = True
            elif first == "religion":
                tmp_class.append(3)
                will_append = True                
            elif first == "politics":
                tmp_class.append(5)
                will_append = True
            elif first == "other":
                if second == "age":
                    will_append = True
                    tmp_class.append(4)
                elif second == "feminist":
                    will_append = True
                    tmp_class.append(1)
    if will_append == True:
        text_list.append(comment)
        class_list.append(tmp_class)
    
# offensive true 고 group 없으면 넘어감 
labeled_kold_df = pd.DataFrame({
    'text' : text_list,
    'class': class_list   
})

print(len(labeled_kold_df))
labeled_kold_df.head()
# kold_df
# 단순혐오
# 성차별
# 정치적 차별
# 종교차별
# 인종차별
```

```python
# 일반적 
# 0. 일반발언 common 
# 1. 성차별 sex
# 2. 인종차별 
# 3. 종교차별 religion
# 4. 연령차별 age
# 5. 정치차별 political
# 6. 혐오욕설 
# 7. 출신차별
# 8. 외모차별
# 9. 성소수자 차별
# 10. 기타혐오 other 
# labeled_beep_df = pd.concat([sex_beep_df, common_beep_df], ignore_index=True)

all_labeled_df = pd.concat([
    labeled_apeach_df, labeled_beep_df, labeled_hate_score_df, 
    labeled_KMHAS_df,labeled_kold_df],
    ignore_index=True
)

print(len(all_labeled_df))
all_labeled_df.to_excel('./all_df.xlsx',index=False)
all_labeled_df.head()
```

```python
real_labels = []
real_texts = []
for idx in range(len(all_labeled_df)):
    labels = all_labeled_df.loc[idx]["class"]
    text = all_labeled_df.loc[idx]["text"]
    labels = list(set(labels))
    
    real_labels.append(labels)
    real_texts.append(text)
    
fixed_labeled_data = pd.DataFrame({
    'text' : real_texts,
    'class': real_labels   
})

fixed_labeled_data.to_excel('./fixed_labeled_data.xlsx',index=False)
```
