```python
import pandas as pd 

labeled_df = pd.read_excel('../1_DataDownload/fixed_labeled_data.xlsx')
print(len(labeled_df))
labeled_df.sample(10)
```

```python
filtered_df = labeled_df[labeled_df['class'].apply(lambda x: len(eval(x)) == 1)]
print(len(filtered_df))
filtered_df.sample(10)
```

```python
filtered_df['class'] = filtered_df['class'].apply(lambda x: eval(x)[0])
filtered_df.sample(10)
```

```python
filtered_df.to_excel('./filtered_data.xlsx',index=False)
```
